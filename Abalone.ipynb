```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif():
    vif=pd.DataFrame()
    vif['Feature name']=x.columns
    vif['Varience Inflation Factor']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    print(vif)
```

```python
calc_vif()
```

- Length and shucked weight has less correlation with target variable with respect to Diameter and Whole weight
- so we drop Length and shucked weight

```python
df.drop(['Length','Shucked weight'],axis=1,inplace=True)
df
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
stdscale=StandardScaler()
x=stdscale.fit_transform(x)
x
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=.20,random_state=45)

lr=LinearRegression()
lr.fit(train_x,train_y)
predlr=lr.predict(test_x)
print('mean_absolute_error=',mean_absolute_error(predlr,test_y))
print('mean_squared_error=',mean_squared_error(predlr,test_y))
print('mean_absolute_error=',np.sqrt(mean_squared_error(predlr,test_y)))
```

```python
lr.score(train_x,train_y)
```

```python
print(r2_score(predlr,test_y))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsRegressor
dtr=DecisionTreeRegressor()
svc=SVC()
knn=KNeighborsRegressor()

model=[dtr,svc,knn]

for m in model:
    m.fit(train_x,train_y)
    pred=m.predict(test_x)
    print("score of ",m,"is")
    print('mean_absolute_error=',mean_absolute_error(pred,test_y))
    
    print('mean_squared_error=',mean_squared_error(pred,test_y))
    
    print('mean_absolute_error=',np.sqrt(mean_squared_error(pred,test_y)))
```

- SVC is performance score is very well among all algorithm.
